```python
import pandas as pd
```

```python
jobreq_data = pd.read_csv("skills.csv")
cv_data = pd.read_csv("candidate_skills.csv")
```

```python
# Function to calculate points based on the description
def calculate_points(candidate_skills, required_skills):
    num_candidate_skills = len(candidate_skills)
    num_required_skills = len(required_skills)
    
    # Calculate the intersection length
    common_skills = len(set(candidate_skills).intersection(set(required_skills)))
    # Determine the denominator (bigger length)
    denominator = max(num_candidate_skills, num_required_skills)
    
    # Calculate the points
    points = common_skills / denominator if denominator != 0 else 0
    return points

# Prepare a list to store the points for each candidate and job requirement
all_points = []

# Iterate over each set of required skills
for req_skills in jobreq_data['skills']:
    # Ensure req_skills is a list (if it is read as a string from CSV)
    req_skills = eval(req_skills) if isinstance(req_skills, str) else req_skills
    
    points = cv_data['skills'].apply(lambda x: calculate_points(eval(x), req_skills) if isinstance(x, str) else calculate_points(x, req_skills))
    all_points.append(points)

# Create a DataFrame from the points list
points_df = pd.DataFrame(all_points).T

# Rename the columns to indicate the job requirement they correspond to
points_df.columns = [f'points_job_{i+1}' for i in range(points_df.shape[1])]

# Combine the candidate skills with the calculated points
cv_data_with_points = pd.concat([cv_data, points_df], axis=1)
```
